```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_excel("dataset_processado.xlsx")
```

```python
dataframe.head()
```

```python
from sklearn import tree
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_validate
from sklearn.metrics.scorer import make_scorer
from sklearn.metrics import confusion_matrix
from functools import reduce
```

```python
res = [] 
```

```python
kf = StratifiedKFold(n_splits=5)
X,Y = dataframe[dataframe.columns[:-1]],dataframe[dataframe.columns[-1]]    
```

```python
start = 5
end = 30
```

```python
for u in range(1,10):
    for i in range(5,30):
        clf = tree.DecisionTreeClassifier(max_depth=u,max_leaf_nodes=i,criterion="entropy")
        sumAcc = []
        for train_index,test_index in kf.split(X,Y):
            xTrain,yTrain = dataframe.iloc[train_index][dataframe.columns[:-1]],dataframe.iloc[train_index][dataframe.columns[-1]]
            xTest,yTest = dataframe.iloc[test_index][dataframe.columns[:-1]],dataframe.iloc[test_index][dataframe.columns[-1]]
            clf.fit(xTrain,yTrain)
            clf.predict(xTest)
            Acc=clf.score(xTest,yTest)
            sumAcc.append(Acc) 
        res.append([np.array(sumAcc).mean(),u,i])
```

```python
results = pd.DataFrame(np.matrix(res),columns=["Acc","Max_depth","Max_leaf_nodes"])
```
