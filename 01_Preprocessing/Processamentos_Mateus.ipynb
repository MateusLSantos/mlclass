```python
n = 0;
for val in lrn['Glucose']:
    if np.isnan(val):
        #print("IsNan: ", n)
        lrn = lrn.drop(index = n);
    n = n+1;

lrn.index = range(len(lrn))
```

```python
n = 0;
for val in lrn['SkinThickness']:
    if val > 80:
        #print("IsNan: ", n)
        lrn = lrn.drop(index = n);
    n = n+1;

lrn.index = range(len(lrn))
```

<h1> Preenchendo com mdia/mediana </h1> 

```python
#Calculando as mdias
n = 0; #Nmero de membros
s = 0; #Soma
for val in lrn['BMI']:
    if not np.isnan(val):
        s = s + val;
        n = n + 1;

meanBMI = round(s/n, 1);

n = 0; #Nmero de membros
s = 0; #Soma
for val in lrn['BloodPressure']:
    if not np.isnan(val):
        s = s + val;
        n = n + 1;

meanBP = round(s/n);

n = 0; #Nmero de membros
s = 0; #Soma
for val in lrn['SkinThickness']:
    if not np.isnan(val):
        s = s + val;
        n = n + 1;

meanST = round(s/n);

n = 0; #Nmero de membros
s = 0; #Soma
for val in lrn['Insulin']:
    if not np.isnan(val):
        s = s + val;
        n = n + 1;

meanIns = round(s/n);


#Ideia: ao invs de usar apenas uma mdia, poderamos usar as mdias estratificadas, uma para cada grupo, depois.
```

```python
#Atribuindo as mdias s clulas vazias

n = 0;
for val in lrn['BMI']:
    if np.isnan(val):
        lrn.at[n, 'BMI'] = meanBMI;
        #lrn.at[n, 'BMI'] = 0; #Preencher com 0 ao invs da mdia
    n = n + 1;

n = 0;
for val in lrn['BloodPressure']:
    if np.isnan(val):
        lrn.at[n, 'BloodPressure'] = meanBP;
        #lrn.at[n, 'BloodPressure'] = 0; #Preencher com 0 ao invs da mdia
    n = n + 1;
    
n = 0;
```

```python
for val in lrn['SkinThickness']:
    if np.isnan(val):
        lrn.at[n, 'SkinThickness'] = meanST;
        #lrn.at[n, 'SkinThickness'] = 0; #Preencher com 0 ao invs da mdia
    n = n + 1;
    
n = 0;
for val in lrn['Insulin']:
    if np.isnan(val):
        lrn.at[n, 'Insulin'] = meanIns;
        #lrn.at[n, 'BMI'] = 0; #Preencher com 0 ao invs da mdia
    n = n + 1;
```
